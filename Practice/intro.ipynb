```python
import numpy as np
import matplotlib.pyplot as plt
import librosa
import librosa.display
import IPython.display as ipd

plt.rcParams.update({'font.size': 16,'axes.grid': True})
SR = 22050

wide = (18, 3)
big = (18, 8)

print(f"{librosa.__version__}")
```

```python
src , sr = librosa.load(librosa.example('trumpet'), sr=SR,mono=True,duration=5.0)
```

```python
plt.figure(figsize=wide)
plt.title('Example waveform')
plt.plot(src)
plt.ylim([-1,1])
```

```python
plt.figure(figsize=wide)
plt.title('Example waveform')
librosa.display.waveshow(src, sr=SR)
plt.ylim([-1,1])
plt.xlim([1, 2])
```

### Spectrograms :

- Short Term Fourier Transformatiom 
- Mel-spectrogram
- Constant-Q Transform

## STFT

```python
n_fft = 512
hop_length = n_fft // 4
stft_complex = librosa.stft(y=src, n_fft=n_fft, hop_length=hop_length)
```

```python
stft = np.abs(stft_complex)
# stft = np.log(np.abs(stft_complex)+0.001)

plt.figure(figsize=wide)
img = plt.imshow(stft)
plt.colorbar(img)
plt.ylabel('bin index\n(lower index -> lower frequency)')
plt.xlabel('time index')
plt.title('Magnitude spectrogram')
plt.grid(False)
```

```python
stft_disrtibution = np.mean(stft, axis=1)
plt.plot(stft_disrtibution)
```

```python
plt.figure(figsize=wide)

stft = np.log(np.abs(stft_complex)+0.001)
plt.imshow(stft, aspect='auto', origin='lower')
```

```python
plt.figure(figsize=wide)

mel_spectrogram = librosa.feature.melspectrogram(y=src, sr=SR, n_fft=n_fft, hop_length=hop_length, n_mels=128)
plt.imshow(mel_spectrogram + stft, aspect='auto', origin='lower')
```
